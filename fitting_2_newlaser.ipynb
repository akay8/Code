Analysing and plotting data for N number of files

```python
## Location of No Insulation Data
path = 'I:/Science/Physics-Photonics/Group/Allan_OPM_MEMS_Fab_and_Clocks/PostDoc_Mainly_RbTPT/TPT_signal_data/Infleqtion_wafer/Infleqtion_Phoenix_wafer/Infleqtion_Phoenix_wafer/Abi/2026-01-13'

## Collect paths of all CSV files from folder into a list
csv_files = glob.glob(os.path.join(path, "*.csv"))
csv_files = csv_files[1:10] ## Only take relevant 9 files
numb_files = len(csv_files) # The number of files

print(csv_files)

''' Key to change '''
deldata = 800 # How much data either side of Lorentz peak is used for fitting
''''''

# Saving all important quantities into empty arrays
ref = np.zeros(numb_files)
fwhm = np.zeros(numb_files)
fwhmerr = np.zeros(numb_files)
cs_arr = np.zeros(numb_files)
rcs_arr = np.zeros(numb_files)
freqp_arr = np.zeros((numb_files,2*deldata))
sigp_arr = np.zeros((numb_files,2*deldata))
popt_arr = np.zeros((numb_files,3))
perr_arr = np.zeros((numb_files,3))
residuals_arr = np.zeros((numb_files,2*deldata))

## Fitting - Loretzian dist.
# 3 params - amplitude A, centre point C, linewidth gamma (FWHM)

def lorentz(xdata,A,C,gamma): 
    x1 = (xdata-C)/(gamma/2)
    return A / (1 + x1**2)


for i in range(numb_files): # some humber of csvs defined above

    ## Take a file and extract data into dataframe
    df = pd.read_csv(csv_files[i], skiprows=11)

    ## Display data frame
    df.head()

    ## extracting time base (x) and amplitude (y) data
    ## ommiting column heading
    x = (df['Second'].to_numpy())[1:] 
    y = (df['Value'].to_numpy())[1:]

    ## Converting data to float
    x = x.astype(float)[1000:9000] #index cutting data to peaks of interest
    y = y.astype(float)[1000:9000]

    ## Inverting signal
    y = np.flip(y)

    plt.plot(x,y)

    # Find Peaks - adjusted params to only find 2 peaks
    p, properties = find_peaks(y, height = -0.02, distance = 150, prominence = 0.02)

    # rescale x and y axes
    ptime = np.array([p[0],p[1]])
    deltat = x[ptime[1]] - x[ptime[0]]
    deltaf = 14.412 #MHz

    print(ptime)

    # Want x axis referenced from the first set of data - save all ref points then only use first one
    ref[i] = x[ptime[1]]

    # zero point is set to centre of largest peak of first data set; freq is in MHz
    freq = -(x - ref[0]) * deltaf / deltat

    # Get rid of y offset
    sig = (y - np.mean(y[0:9]))*1e3 #NB units are mV

    ## Chop data to focus on biggest peak - modify for each data set with ptime[1]
    first = ptime[1]-deldata
    last = ptime[1]+deldata
    sigp = sig[first:last]
    freqp = freq[first:last]

    ## Save data used for fits
    freqp_arr[i,:] = freqp
    sigp_arr[i,:] = sigp
    
    # Initial guess for lorentz params
    guess = np.array([90,0,1])

    popt,pcov = curve_fit(lorentz,freqp,sigp,p0=guess)
    perr = np.sqrt(np.diag(pcov))
    popt[2] = np.abs(popt[2]) # have positive fwhms

    # Save popt,perr
    popt_arr[i,:] = popt
    perr_arr[i,:] = perr

    ## Save fwhm and errs into arrays
    fwhm[i] = popt[2]
    fwhmerr[i] = perr[2]

    print('The full width half maximum of file {} is {} pm {} Mhz'.format(i+1,popt[2],perr[2]))

    ## Chi Squared

    # degrees of freedom = size of dad set - fit params
    dof = np.size(freqp) - np.size(popt) 

    # chi squared function
    def chi_squared(xdata,ydata,f,params):
        return np.sum(((ydata - f(xdata,*params))**2) / f(xdata,*params))

    # Calculate cs and reduced cs
    cs = chi_squared(freqp,sigp,lorentz,popt)
    red_cs = cs / dof
    print('The Chi Squared is {}'.format(cs))
    print('The Reduced Chi Squared is {}'.format(red_cs))

    ## Save chi squareds
    cs_arr[i] = cs
    rcs_arr[i] = red_cs

    # Residuals
    def residuals(xdata,ydata,f,params):
        return (ydata - f(xdata,*params))

    resd = residuals(freqp,sigp,lorentz,popt)
    residuals_arr[i,:] = resd
```

Plotting with saved data from 10 files

```python
# Plotting
fig, (ax1,ax2) = plt.subplots(2,1, height_ratios = [4,1], sharex='all')

ax1.grid(True)
ax2.grid(True)
ax2.set_xlabel('\n Frequency / MHz')
ax1.set_ylabel('Voltage / mV \n')
ax2.set_ylabel('Residuals / mV \n')

blues = colormaps['Blues'].resampled(256)
colors = blues(np.linspace(0.4, 1, numb_files))

for k in range(numb_files):

    ax1.set_title('Lineshape of all files - Uninsulated PMT')
    ax1.scatter(freqp_arr[k,:],sigp_arr[k,:], s = 1, color = colors[k])
    ax1.plot(freqp_arr[k,:],lorentz(freqp_arr[k,:],*popt_arr[k,:]), color = 'k')
    #ax1.legend()

    ax2.scatter(freqp_arr[k,:],residuals_arr[k,:],s=1, color = colors[k])

fwhm_mean = np.mean(fwhm)

fig1, ax = plt.subplots()
ax.errorbar(np.arange(1,numb_files+1), fwhm, yerr = perr_arr[:,2], ls = '', marker = 'o', ms = 4, color = 'k')
ax.set_xlabel('File Number')
ax.set_ylabel('Linewidth (FWHM) / MHz')
ax.axhline(fwhm_mean, label = 'Mean')
ax.legend()

print(fwhm_mean)
```
